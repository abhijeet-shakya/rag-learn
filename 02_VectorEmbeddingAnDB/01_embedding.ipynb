### What is an embedding?
- Computre can't understand text they only understand numbers or math. An embedding is a translation layer. It takes text and convert in to list of numbers(a vector). The numbers are not random they represent meaning of text in vector space. So an embedding is the simply capture the semantic meaning of the text into coordinates so the computer can understand or do math on it.

- Different different models have different dimensions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simplified 2D example (real embeddings have hundreds of dimension)

word_embeddings = {
    "cat": [0.8,0.6],
    "kitten": [0.75, 0.65],
    "dog": [0.7, 0.3],
    "puppy":[0.65, 0.35],
    "car":[-0.5, 0.2],
    "truck": [-0.45, 0.15]
}
```

```python
fig, ax = plt.subplots(figsize=(8,6))

for word, coords in word_embeddings.items():
    ax.scatter(coords[0], coords[1], s=100)
    ax.annotate(word, (coords[0], coords[1]), xytext= (5,5), textcoords = 'offset points')

ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_title('Simlified word embedding in 2d space')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

### Measuring Similarity

```python
def cosine_similarity(vec1, vec2):
    """
    Cosine similarity measure the angle between tow vectores.
    - Result close to 1 : Very similar
    - Resulte close to 0 : Not related
    - Result close to -1 : Opposite meanings
    """
    dot_product = np.dot(vec1, vec2)
    norm_a = np.linalg.norm(vec1)
    norm_b = np.linalg.norm(vec2)
    return dot_product/(norm_a* norm_b)
```

```python
# Example
cat_vector = [0.8, 0.6, 0.3]
kitten_vector = [0.75, 0.65, 0.35]
car_vector = [-0.5, 0.2,0.1]

cat_ktten_similarity = cosine_similarity(cat_vector, kitten_vector)
print(cat_ktten_similarity)
```

```python
cosine_similarity(cat_vector, car_vector)
```
